```python
data_folder = 'mnm2_processed_data/train'
```

```python
def ShowSlicesAcrossAxis(patient, phase, num, type):
  if type=='images':
    data_load = nib.load('{}/images/{}_SA_{}_{}.nii.gz'.format(data_folder, patient, phase, num)).get_fdata()
  else:
    data_load = nib.load('{}/labels/{}_SA_{}_gt_{}.nii.gz'.format(data_folder, patient, phase, num)).get_fdata()
  plt.figure(figsize=(20, 20))
  for i in range(data_load.shape[2]):
    plt.subplot(1,data_load.shape[2],i+1)
    plt.imshow(data_load[:,:,i], 'gray', interpolation='none')
  plt.show()

def ShowSlicesAcrossAxis_annotated(patient, phase, num):
  img_data = nib.load('{}/images/{}_SA_{}_{}.nii.gz'.format(data_folder, patient, phase, num)).get_fdata()
  lab_data = nib.load('{}/labels/{}_SA_{}_gt_{}.nii.gz'.format(data_folder, patient, phase, num)).get_fdata()
  plt.figure(figsize=(20, 20))
  for i in range(img_data.shape[2]):
    plt.subplot(1,15,i+1)
    plt.imshow(img_data[:,:,i], 'gray', interpolation='none')
    plt.imshow(lab_data[:,:,i], 'jet', alpha=0.2, interpolation='none')
  plt.show()
```

```python
ShowSlicesAcrossAxis('001', 'ED', '0011', 'images')
```
